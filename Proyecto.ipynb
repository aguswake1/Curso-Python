```python
!wget 'https://covid.ourworldindata.org/data/ecdc/full_data.csv'
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
```

```python

"""Comienzo a desarrollar el primer punto del TP integrador"""

plt.rcParams['figure.figsize'] = (17,10.2)  # Establezco las medidas de los gráficos (x, y)
csv = pd.read_csv('full_data.csv', header=0)

country = input(str("Ingrese el nombre de un pais (en ingles) \n"))

while (country.isalpha() is not True):
  country = input(str("Ingrese un nombre de pais válido: "))
else:
  pass

search_country_item = csv[csv.iloc[:,1] == country]  # El metodo iloc busca por indice de fila , columna. En este caso columna 1 (location)
total_cases = search_country_item['total_cases']  # En el archivo casos meto toda la columna "new_cases" ya filtrada
total_deaths = search_country_item['total_deaths']
dates = search_country_item['date']  # En el archivo fechas meto toda la columna "date" ya filtrada

plt.plot(dates, total_cases, 'r-', label='Total de infectados en ' + country)
plt.plot(dates, total_deaths, 'k-', label='Total de fallecidos en ' + country)
plt.xticks(rotation=90)
plt.legend(loc='upper left')
plt.show()
```

```python

"""Comienzo a desarrollar el cuarto punto del TP integrador"""  # Crea un archivo en la dirección almacenada del proyecto

i = 0
archivo = pd.read_csv('full_data.csv', header = 0 )
pais = (archivo[archivo.iloc[:,1] == "China"])  # Filtro por un pais cualquiera solo para obtener el total de dias
fechas = pais['date']  # Obtengo la lista de todos los dias
grabador = pd.ExcelWriter('Salida.xlsx')  # Creo el archivo .xlsx

for i in range (len(fechas)):  # Un for que corre dia por dia
    paises = (archivo[archivo.iloc[:,0] == fechas.iloc[i]])  # Obtengo los valores para cada pais solo del dia especifico
    paises = paises.sort_values(by = ['total_cases'],ascending=[False])  # Los ordeno de mayor a menor
    hoja = 'hoja '+ str(i + 1)  # Creo el nombre de la hoja del excel para ese dia
    paises.to_excel(grabador, sheet_name = hoja, index=False)  # Guardo en la hoja los datos

grabador.save()  # Guardo el archivo
grabador.close()  # Cierro el archivo
```
